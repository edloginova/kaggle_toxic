```python
import pickle
import pandas as pd
import sklearn
import numpy as np
```

```python
MODEL_PATH = "../models/"
DATA_PATH = "../data/"
```

```python
model_names = ['keras_lstm_fancy_classification_model', 'keras_lstm_classification_model', 'evgeniya/rf_joint', 'cnn_512filters']
preds = []
for model_name in model_names:
    preds.append(pd.read_csv("../submissions/" + model_name + ".csv"))
```

```python
list_classes = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
```

## average

```python
average_pred = preds[0].copy(deep = True)
```

```python
for class_name in list_classes:
    average_pred[class_name] = np.mean([pred[class_name].values for pred in preds], axis = 0)
```
