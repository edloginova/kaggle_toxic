```python
import numpy as np
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
import pickle
```

```python
DATA_PATH = "../data/"
train = pd.read_csv(DATA_PATH + "train.csv")
train = train.sample(frac=1)
list_classes = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
y = train[list_classes].values
```

```python
y_train = pickle.load(open("../submissions/TRAINdeep_dense_model_f2_score.hdf5.p", "rb"))
```

```python
y_train.shape
```

```python
roc_aucs = []
accuracy_scores = []
for i in range(0, y_train.shape[1]):
    roc_aucs.append(roc_auc_score(y[:, i], y_train[:, i]))
    accuracy_scores.append(accuracy_score(y[:, i], np.round(y_train[:, i])))
```
